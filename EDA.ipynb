```python
telco_data = telco_base_data.copy() # copy for manipulation
```

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce') # convert total : numeric
telco_data.isnull().sum()
```

```python
telco_data.loc[telco_data ['TotalCharges'].isnull() == True] # 11 missing values, removed
```

```python
# Removing missing values (drop rows that have at least one missing (NaN) value in any of their columns)
telco_data.dropna(how = 'any', inplace = True)
```

```python
# Get the max tenure
# Divide customers into bins based on tenure < 12 months: group of 1-12, 1-2 years, 2-3 years etc.
print(telco_data['tenure'].max()) #72
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]
telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

```python
# drop column customerID and tenure
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
```

```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0) # churn into binary
```

```python
telco_data.head()
```

```python
telco_data_dummies = pd.get_dummies(telco_data) # one hot encoder : categorical -> numerical (19 columns to 50 columns)
telco_data_dummies.head()
```

```python
plt.figure(figsize=(20,8))
telco_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')

# Churn in Month to month contracts, No online security, No Tech support, First year of subscription and Fibre Optics Internet ; high churn

# Churn in Long term contracts, Subscriptions without internet service and The customers engaged for 5+ years ; low churn

# Gender, Phone Service, Multiple Lines : No impact on churn
```

```python
telco_data_dummies.to_csv('tel_churn.csv') # uploading
```
